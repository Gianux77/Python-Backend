COMO FUNCIONA LA WEB

_mediante una arquitectura de cliente-servidor.

_estas computadoras intercambian paquetes de informacion entre 2 direcciones. A esto lo llamamos el protocolo TCP/IP.

los mensajes que se envia el cliente y el servidor usan el protocolo HTTP para estructurarse.

El servidor actua como un proveedor de servicios, y podemos decir que comparte sus recursos de computo con los clientes.

para que todo pueda funcionar se usan protocolos de red, en el caso de la web se usa el modelo TCP/IP.

QUE ES UN PROTOCOLO DE RED.

Un protocolo es un conjunto de reglas y normas preestablecidas, que para el caso de los protocolos de red, sirven para transmitir datos entre dos dispositivos.

Â¿Que busca determinar?

-Como esructurar los datos.
-El formato de los datos.
-Que informacion enviar.
-Que hacer en caso de error.

Estos protocolo trabajan en colaboracion, como en la fabrica en edonde cada protocolo reciben datos,los procesa, y los pasa al siguiente protocolo, asi hasta llegar a su destino.

MODELO TCP/IP

esun modelo que agrupa a distintos protocolos, y que se utiliza para establecer comunicacion entre dos dispositivos a traves de internet.

se agrupan en 5 capas:

_capa fisica

_capa de enlace de datos(Data Link)

_capa de red

_capa de transporte

_capa de aplicacion


Diagrama que puedes disinguir los protocolos y que lugar en la capa ocupan.

HTTP,FTP,SMTP _(5) APPLICATION

TCP,UDP       _(4) TRANSPORT

IP,ROUTERS    _(3) NETWORK

ETHE,SWITCHES _(2) DATA LINK

CABLES,NIC    _(1) PHYSICAL




CAPA FISICA

hace referencia al medio fisico por el cual se movera la informacion, esto incluye los cables ethernet, los cables de fibra optica, entre otros.

CAPA DATA LINK

funciona de intermediario entre la capa de red y la capa fisica, su trabajo es procesar los paquetes de datos de las capas superiores, y transformarla en bits que puedan transmitirse por la capa fisica.

CAPA RED

establece el camino entre un dispositivo y otro, asi como el de transformar los datos que vienen de capas superiores en paquetes.

CAPA DE TRANSPORTE

supervisa el camino de la informacion, ej, ip, enruta el mensaje hacia el destinatario, sin que exista una confirmacion de que el mensaje ha llegado.

CAPA DE APLICACION

es el enlace con la aplicacion con la que interactua el usuario, y establece el mensaje para pasar a la capa de transporte.


para enviar informacion utilizara uno de dos principales protocolos de transpore, TCP Y UDP quienes se encargan de supervisar el envio y comunicar a ambos dispositivos.

PROTOCOLO HTTP(Hyper Text Transfer Protocol)

es el protocolo de la web, se ubica en la capa de aplicacion por lo que podemos pensar en este como la puerta para que nuestro codigo se comunique con el internet y podamos intercambiar datos.

principales caracterisica:

_Es simple: puede ser leido por humanos

_Es extendible: nueva funcionalidad puede ser implementada usando encabezados.

_No tiene estado, pero tiene sesiones: cada mensaje es independiente.

_elegir una capa de transporte de confianza.

SOLICITUD

_version del protocolo HTTP
_URL
_metodo HTTP
_encabezados HTTP
_cuerpo HTTP(opcional)


URL

son las siglas de (uniform resource locator) que se puede traducir como identificador de recurso uniforme.
las URLs nos permiten identificar los recursos de internet.

Metodo HTTP

identificador busca indicarle al servidor que accion espera que realiza con el recurso de la URL.

los navegadores soportan dos de estos: GET y POST, aunque el protocolo indica otro mas.

Encabezados HTTP

son datos organizados en formato clave/valor arbitrarios.
estos nos permiten extender la funcionalidad del protocolo.

CUERPO HTTP

el cuerpo de la peticion puede contener la informacion adicional que enviamos al servidor.

esta informacion puede enviarse en distinos formatos apoyandonos del encabezado content_type.

CODIGO DE ESTADO

le permite saber al cliente cual fue el resultado de la peticion, ejemplo, si se completo con exito, o algo mas paso.

100 informativo
200 exito
300 redireccion
400 error cliente
500 error servidor
